```python
def counting_rationale_len(df):
    rationale_dict = {}
    rationale_len = []
    for doc_evidences in df['evidences']:
        for evidence in doc_evidences:
            a = evidence[0]['end_token']-evidence[0]['start_token']
            b = len(evidence[0]['text'].split())
            assert (a == b)
            rationale_len.append(a)
            doc_id = evidence[0]['docid']
            if doc_id not in rationale_dict.keys():
                rationale_dict[doc_id] = a
            else:
                rationale_dict[doc_id] += a
    return rationale_dict
```

```python
def counting_rationale_len(df):
    rationale_len = []
    for doc_evidences in df['evidences']:
        l = 0
        for evidence in doc_evidences:
            a = evidence[0]['end_token']-evidence[0]['start_token']
            b = len(evidence[0]['text'].split())
            assert (a == b)
            l += a
        rationale_len.append(l)
    return rationale_len
```

```python
rationale_dict = {**counting_rationale_len(train), **counting_rationale_len(test), **counting_rationale_len(val)}
rationale_len = list(rationale_dict.values())
sum(rationale_len)/len(rationale_len)
```

```python
def file_len(file):
    file = open(file, 'r')   
    f_len = 0
    for line in file.readlines():
        f_len += len(line.rstrip().split())
    return f_len
```

```python
f_dict = {}
data_dir = directory + 'docs'
for filename in os.listdir(data_dir):
    f = os.path.join(data_dir, filename)
    if os.path.isfile(f):
#         f_lens.append(file_len(f))
        f_dict[filename] = file_len(f)
f_lens = list(f_dict.values())
sum(f_lens)/len(f_lens)
```
